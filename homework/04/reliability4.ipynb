```python
from scipy.stats import norm, expon
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 9
sigma = 0.05 * mu
normdist = norm(mu, sigma)


def Pff(t):
    return 1 - normdist.cdf(t)
```

```python
p8h = 1 - normdist.cdf(8)
print('p8h: ' + str(p8h))

p8h30m = 1 - normdist.cdf(8.5)
print('p8h30m: ' + str(p8h30m))

p8h_30m = p8h30m / p8h
print('p8h_30m: ' + str(p8h_30m))

Q8h_30m = p8h - p8h30m
print('Q8h_30m: ' + str(Q8h_30m))

T1 = quad(Pff, 8.5, float("inf"))/Pff(8.5)
T1 = [round(T1[0], 6), round(T1[1], 10)]

L8h30m = normdist.pdf(8.5) / (1 - normdist.cdf(8.5))

Lambda = normdist.pdf(9) / (1 - normdist.cdf(9))
print('T1: ' + str(T1))
print('L8h30m: ' + str(L8h30m))
print('Lambda: ' + str(Lambda))
```

```python
x = np.linspace(0.1, 10, 100)
y = normdist.pdf(x)
y1 = expon.pdf(x, scale=18)


plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.plot(x, y1, 'r-', linewidth=2)
plt.title('Нормальное распределение')
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0.1, 10, 100)
y = 1. - normdist.cdf(x)
y1 = 1 - expon.cdf(x, scale=18)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.plot(x, y1, 'r-', linewidth=2)
plt.title('Нормальное распределение')
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0.1, 10, 100)
y = normdist.pdf(x) / normdist.sf(x)
y1 = expon.pdf(x, scale=18) / expon.sf(x, scale=18)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.plot(x, y1, 'r-', linewidth=2)
plt.title('Нормальное распределение')
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.grid(True)
plt.show()
```
